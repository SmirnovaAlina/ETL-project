```python
#Extract

#Dependencies

import pandas as pd
import requests
import json 
from sodapy import Socrata

client = Socrata("data.sfgov.org", None)
results = client.get("wwmu-gmzc", limit=2000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
results_df
```

```python
file = "Resources/database.csv"

df_awards = pd.read_csv(file)
df_awards.head()
```

```python
file = "Resources/movie_metadata.csv"

movie_df = pd.read_csv(file)
```

```python
#Transform
new_df = results_df.merge(df_awards, left_on='title',right_on='Film', how ='left')
new_df 
```

```python
new_df2 = new_df.dropna(subset=['Film'])
new_df2
```

```python
movie_df2=movie_df[['gross', 'budget','movie_title','movie_facebook_likes']]
movie_df2.head()
```

```python
movie_df.columns
```

```python
combined_df = new_df2.merge(movie_df2, left_on='title', right_on='movie_title', how ='left')
combined_df
```

```python
combined_df.dtypes
```

```python
final_df = combined_df.drop(columns=['actor_1','actor_2','actor_3', 'actor_3', 'fun_facts','title','Year','Ceremony','Winner','gross','budget', 'movie_title', 'movie_facebook_likes'])
final_df=final_df.reset_index()
final_df
```

```python
#create actor_movie_awards 

grouped_actor_movie_awards_df = final_df.groupby(['Name','Film'])
grouped_actor_movie_awards_df
```

```python
actor_movie_awards_df = pd.DataFrame({'movie_name': grouped_actor_movie_awards_df['Film'].unique().str.get(0), 
                                       'actor_name': grouped_actor_movie_awards_df['Name'].unique().str.get(0),
                                       'count_awards_by_actor_movie': grouped_actor_movie_awards_df['Award'].count()})
actor_movie_awards_df.head()
```

```python
actor_movie_awards_df=actor_movie_awards_df.reset_index(drop=True)
actor_movie_awards_df.head()
```

```python
#create movies_awards 

grouped_movies_df = final_df.groupby(['Film'])
grouped_movies_df 
```

```python
movies_df = pd.DataFrame({'movie_name': grouped_movies_df['Film'].unique().str.get(0), 
                                       'director': grouped_movies_df['director'].unique().str.get(0),
                                       'location': grouped_movies_df['locations'].unique().str.get(0),
                                       'production_company': grouped_movies_df['production_company'].unique().str.get(0),
                                       'release_year': grouped_movies_df['release_year'].unique().str.get(0),
                                       'writer': grouped_movies_df['writer'].unique().str.get(0),
                                       'count_awards_by_movie': grouped_movies_df['Award'].count()})
movies_df.head()
```

```python
movies_df=movies_df.reset_index(drop=True)
movies_df
```

```python
#create actors_awards 

grouped_actors_df = final_df.groupby(['Name'])
grouped_actors_df 
```

```python
actors_df  = pd.DataFrame({ 'actor_full_name': grouped_actors_df['Name'].unique().str.get(0)})
                            
actors_df.head()
```

```python
actors_df=actors_df.reset_index(drop=True)
actors_df
```

```python
actors_df['actor_name']=actors_df['actor_full_name'].str.rsplit(n=1)
actors_df['actor_lastname']=actors_df['actor_name'].str.get(1)
actors_df['actor_firstname']=actors_df['actor_name'].str.get(0).str.split(n=1).str.get(0)
actors_df['actor_middlename']=actors_df['actor_name'].str.get(0).str.split(n=1).str.get(1)
actors_df.head()
```

```python
actors_df=actors_df[['actor_full_name', 'actor_lastname','actor_firstname','actor_middlename']]
actors_df
```
